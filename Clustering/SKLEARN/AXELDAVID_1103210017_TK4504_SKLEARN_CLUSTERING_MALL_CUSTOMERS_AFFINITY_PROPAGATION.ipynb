```python
df.rename(columns={'Genre': 'Gender'}, inplace=True)
df = pd.get_dummies(df, columns=['Gender'])
df['Gender'] = df['Gender_Male'].astype(int)
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
X = scaled_data

af = AffinityPropagation(preference=-50, random_state=0).fit(X)
cluster_centers_indices = af.cluster_centers_indices_
labels = af.labels_

n_clusters_ = len(cluster_centers_indices)

print("Estimated number of clusters: %d" % n_clusters_)
print("Silhouette Coefficient: %0.3f" % metrics.silhouette_score(X, labels, metric="sqeuclidean"))

plt.close("all")
plt.figure(figsize=(10, 8))

colors = plt.cm.viridis(np.linspace(0, 1, n_clusters_))

for k, col in zip(range(n_clusters_), colors):
    class_members = labels == k
    cluster_center = X[cluster_centers_indices[k]]
    plt.scatter(
        X[class_members, 0], X[class_members, 1], color=col, marker=".", label=f'Cluster {k}'
    )
    plt.scatter(
        cluster_center[0], cluster_center[1], s=200, color=col, marker="o", label=f'Cluster {k} center'
    )
    for x in X[class_members]:
        plt.plot(
            [cluster_center[0], x[0]], [cluster_center[1], x[1]], color=col
        )

plt.title("Estimated number of clusters: %d" % n_clusters_)
plt.legend()
plt.show()
```

```python
af = AffinityPropagation(preference=-50, random_state=0)
cluster_labels = af.fit_predict(df)

lle = LocallyLinearEmbedding(n_components=2, n_neighbors=6, random_state=0)
embedding = lle.fit_transform(df)

cluster_centers_indices = af.cluster_centers_indices_
cluster_centers = embedding[cluster_centers_indices]

plt.figure(figsize=(10, 8))
for cluster_center in cluster_centers:
    plt.scatter(cluster_center[0], cluster_center[1], marker='o', color='black', s=100)
for i in range(len(embedding)):
    plt.text(embedding[i, 0], embedding[i, 1], str(cluster_labels[i]), color=plt.cm.viridis(cluster_labels[i] / len(cluster_centers)))
for i in range(len(embedding)):
    for j in range(i + 1, len(embedding)):
        if cluster_labels[i] == cluster_labels[j]:
            plt.plot([embedding[i, 0], embedding[j, 0]], [embedding[i, 1], embedding[j, 1]], color=plt.cm.viridis(cluster_labels[i] / len(cluster_centers)), alpha=0.3)
plt.title('Affinity Propagation Clustering with Locally Linear Embedding')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.colorbar(label='Cluster')
plt.show()
```
