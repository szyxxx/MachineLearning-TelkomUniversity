```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import drive
import pandas as pd

drive.mount('/content/drive')

link = "/content/anime.csv"
dataset = pd.read_csv(link)

print(dataset)
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=dataset.select_dtypes(include=['int64', 'float64']), orient="h", palette="Set2")
plt.title("Boxplot of Numeric Features")
plt.xlabel("Value")
plt.ylabel("Feature")
plt.show()
```

```python
for column in dataset.select_dtypes(include=['float64', 'int64']):
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=dataset[column])
    plt.title(f'Boxplot of {column}')
    plt.show()
```

```python
print("Statistik Fitur:")
(dataset.describe())
```

```python
sns.pairplot(dataset.select_dtypes(include=['float64', 'int64']))
plt.show()
```

```python
plt.figure(figsize=(14, 10))

for i, col in enumerate(dataset.select_dtypes(include=['int64', 'float64'])):
    plt.subplot(3, 3, i + 1)
    sns.histplot(data=dataset, x=col, kde=True)
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 10))

for i, col in enumerate(dataset.select_dtypes(include=['int64', 'float64'])):
    plt.subplot(3, 3, i + 1)
    sns.violinplot(data=dataset, y=col)
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(100, 60))
plt.scatter(dataset['title'], dataset['score'], alpha=0.5)
plt.title('Scatter Plot antara Judul dan Rating')
plt.xlabel('Judul')
plt.ylabel('Rating')
plt.grid(True)
plt.show()
```

```python
correlation_matrix = dataset.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Korelasi antar Fitur dalam Dataset')
plt.show()
```
